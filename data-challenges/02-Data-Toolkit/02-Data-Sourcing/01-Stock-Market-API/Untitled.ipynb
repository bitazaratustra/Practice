```python
import pandas as pd
import requests


try:
    response = requests.get('https://iex.lewagon.com/stable/stock/aapl/chart/3m', timeout=10).json()
    
except:
    raise ValueError('Problemas al consultar la API')
```

```python
response
```

```python
apple_df = pd.DataFrame(response)
apple_df
```

```python
apple_df['date'] = pd.to_datetime(apple_df['date'], format="%Y-%m-%d")
```

```python
apple_df.set_index('date', inplace=True)
apple_df
```

```python
apple_df['close'].plot()
```

```python
apple_df[['open', 'close', 'high', 'low']].plot()
```

```python
apple_df[['open', 'close', 'high', 'low']].plot(figsize=(12,4))
```
