```python
urls = ['https://iex.lewagon.com/stable/stock/amzn/chart/1y','https://iex.lewagon.com/stable/stock/meta/stats','https://iex.lewagon.com/stable/stock/aapl/financials', 'https://iex.lewagon.com/stable/stock/tsla/news/last/1', 'https://iex.lewagon.com/stable/stock/market/sector-performance']
   
```

```python
import pprint

for url in urls:
    results = requests.get(url).json()
    pp = pprint.PrettyPrinter(indent=4, width=41, compact=True)
    pp.pprint(results)
```

```python
dfs = []

for url in urls:
    # Make the request and get JSON response
    response = requests.get(url)
    results = response.json()
    
    # Print out the results to inspect its structure
    print("JSON Response:", results)
    
    # Append DataFrame to dfs
    try:
        dfs.append(pd.DataFrame(results))
    except Exception as e:
        print("Error:", e)

# After inspecting the structure of results and handling any issues, proceed with further processing or debugging
```
